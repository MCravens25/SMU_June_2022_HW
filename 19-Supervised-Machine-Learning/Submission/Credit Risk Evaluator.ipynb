# Prepare our Training Data

### String Columns first

```python
#Loop through the coulumns of the Train_df and bring back the columns that are object type
for col in train_df.columns:
     if train_df[col].dtype == object:
            print(col)
```

```python
#Loop through the columns of the Train_df and bring back the columns that are object type excluding our target
str_cols = []

for col in train_df.columns:
    if (train_df[col].dtype == object) & (col != "loan_status"):
        print(col)
        str_cols.append(col)
        
df_str = train_df.loc[:, str_cols]
df_str.head()
```

```python
#Print the value make up of our columns so we can see if we want to one line encode
for col in df_str.columns:
    print(col)
    print(df_str[col].value_counts())
    print()
```

```python
#Payment plan is useless because all the values are the same
# debt_settlement_flag only 5 columns that are a yes so not really going to hold a correlation to our target
#Drop these rows
df_str.drop(["pymnt_plan", "debt_settlement_flag"], axis=1, inplace=True)
```

```python
# label encode "initial_list_status", "application_type", "hardship_flag" because there is only 2 categories
df_le = df_str.loc[:, ["initial_list_status", "application_type", "hardship_flag"]]
df_le["initial_list_status"] = [1 if x == "f" else 0 for x in df_le.initial_list_status]
df_le["application_type"] = [1 if x == "Joint App" else 0 for x in df_le.application_type]
df_le["hardship_flag"] = [1 if x == "Y" else 0 for x in df_le.hardship_flag]

df_le.head()
```

```python
#One line encode "home_ownership", "verification_status" because there are more than 2 categories
df_oh = pd.get_dummies(df_str.loc[:, ["home_ownership", "verification_status"]])
df_oh.head()
```

### Num Columns Next

```python
#Loop through the columns of the Train_df and bring back the columns that are Not object type and excluding our target
num_cols = []

for col in train_df.columns:
    if (train_df[col].dtype != object) & (col != "loan_status"):
        print(col)
        num_cols.append(col)
        
df_num = train_df.loc[:, num_cols]
df_num.head()
```

```python
# Unscaled Data
# scaler = StandardScaler().fit(df_num)
# df_scale = pd.DataFrame(scaler.transform(df_num), columns=df_num.columns)
df_ns = df_num.copy()
df_ns.head()
```

```python
#Combine our scale data and our oneline encoded df
df_train = pd.concat([df_ns, df_oh, df_le], axis=1)
#Add back in our loan_status target
df_train["loan_status"] = train_df.loan_status
#Change the Target column into a boolean column type with 1 and 0
df_train["loan_status"] = [1 if x == "high_risk" else 0 for x in df_train.loan_status]
df_train.head()
```

```python
#NAN columns because of 1 value for the entire column

corrs = abs(df_train.corr())["loan_status"].sort_values(ascending=False)
corrs          
```

```python
corrs.index
```

```python
#Drop the columns with less than a .01 correlation except for the columns associated with our one line encoded columns

cols_to_drop = ['num_sats', 'open_acc',
       'num_tl_90g_dpd_24m', 'delinq_amnt', 'acc_now_delinq', 'num_tl_30dpd',
       'annual_inc', 'chargeoff_within_12_mths', 'total_il_high_credit_limit',
       'total_bal_ex_mort', 'tot_coll_amt', 'pub_rec_bankruptcies',
       'pub_rec', 'total_acc', 'num_il_tl', 'recoveries',
       'collection_recovery_fee', 'policy_code', 'num_tl_120dpd_2m',
       'tax_liens']

cols_to_keep = [x for x in df_train if x not in cols_to_drop]
df_train = df_train.loc[:, cols_to_keep]
df_train.head()
```

# Prepare our Test Data

```python
df_test = test_df.copy()

df_str_test = df_test.loc[:, str_cols]

# only 1 category essentially...
df_str_test.drop(["pymnt_plan", "debt_settlement_flag"], axis=1, inplace=True)

# label encode
df_le_test = df_str_test.loc[:, ["initial_list_status", "application_type", "hardship_flag"]]
df_le_test["initial_list_status"] = [1 if x == "f" else 0 for x in df_le_test.initial_list_status]
df_le_test["application_type"] = [1 if x == "Joint App" else 0 for x in df_le_test.application_type]
df_le_test["hardship_flag"] = [1 if x == "Y" else 0 for x in df_le_test.hardship_flag]

df_le_test.head()
```

```python
df_oh_test = pd.get_dummies(df_str_test.loc[:, ["home_ownership", "verification_status"]])
df_oh_test.head()
```

```python
df_num_test = df_test.loc[:, num_cols]
#Do not rescale the test data just fit the previous scaler to the test data
#Keep the Test set Unscaled as well
# df_scale_test = pd.DataFrame(scaler.transform(df_num_test), columns=df_num_test.columns)
df_ns_test = df_num_test.copy()
df_ns_test.head()
```

```python
df_test = pd.concat([df_ns_test, df_oh_test, df_le_test], axis=1)

df_test["loan_status"] = test_df.loan_status


df_test["loan_status"] = [1 if x == "high_risk" else 0 for x in test_df.loan_status]
df_test.head()
```

```python
df_test = df_test.loc[:, cols_to_keep]
df_test.head()
```

```python
#Drop our target and our our onelinecoded columns
X_train = df_train.drop(["loan_status", "home_ownership_ANY", "verification_status_Not Verified"], axis=1).values
y_train = df_train["loan_status"].values

X_test = df_test.drop(["loan_status", "home_ownership_ANY", "verification_status_Not Verified"], axis=1).values
y_test = df_test["loan_status"].values
```

```python
def evaluateModel(model, X_train, y_train, X_test, y_test):
    # FIT THE MODEL
    model.fit(X_train, y_train)

    # EVALUATE the MODEL
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)

    # PRINT the METRICS
    print("TRAINING SET")

    class_report_train = classification_report(y_train, train_preds)
    conf_matrix_train = confusion_matrix(y_train, train_preds)

    print(class_report_train)
    print(conf_matrix_train)

    print()

    # PRINT the METRICS
    print("Testing SET")

    class_report_test = classification_report(y_test, test_preds)
    conf_matrix_test = confusion_matrix(y_test, test_preds)

    print(class_report_test)
    print(conf_matrix_test)
    
    y_pred_proba = model.predict_proba(X_test)[::,1]
    fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)

    plt.figure(figsize=(10,6))
    plt.plot(fpr,tpr,label="auc="+str(auc))
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve Test Set")
    plt.legend(loc=4)
    plt.show()
    
    return(model) 
```

### Hypothesis:
#### SInce the data is unscaled right now the Random forest will perform slightly better than the logistic regression because the logistic regression probablitlity will not be as accuarte as it would be with the data scaled.
#### However the random forest will not project that much better a perdiction model because we are selecting random variables and hoping for a strong correlation

```python
# Initialize the model
lr = LogisticRegression()
lr = evaluateModel(lr, X_train, y_train, X_test, y_test)
```

```python
#Drop our target and our our onelinecoded columns
X_train = df_train.drop(["loan_status"], axis=1).values
y_train = df_train["loan_status"].values

X_test = df_test.drop(["loan_status"], axis=1).values
y_test = df_test["loan_status"].values
```
